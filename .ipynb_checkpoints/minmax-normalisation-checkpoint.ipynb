```python

# Define a list of features to check for outliers
features = ['moisture_cheek', 'moisture_forearm', 'moisture_forehead', 'shininess_cheek', 'shininess_forehead', 'evenness_cheek', 'evenness_forearm', 'evenness_red_cheek', 'evenness_red_forearm', 'evenness_red_forehead', 'ita_cheek', 'ita_forearm', 'ita_forehead', 'melanin_cheek', 'melanin_forearm', 'melanin_forehead', 'vessels_den_cheek', 'vessels_den_forearm', 'vessels_den_forehead', 'skin_texture_cheek', 'skin_texture_forearm', 'skin_texture_forehead', 'fine_lines_cheek', 'fine_lines_forearm', 'fine_lines_forehead', 'deep_lines_forehead', 'cheek_age_dif', 'forearm_age_dif', 'forehead_age_dif', 'G_porph_L_MEAN_cheek', 'G_porph_L_MEAN_forehead', 'G_porph_NUMBER_cheek', 'G_porph_NUMBER_forehead', 'R_porph_L_MEAN_cheek', 'R_porph_L_MEAN_forehead', 'R_porph_NUMBER_cheek', 'R_porph_NUMBER_forehead', 'Y_porph_L_MEAN_cheek', 'Y_porph_L_MEAN_forehead', 'Y_porph_NUMBER_cheek', 'Y_porph_NUMBER_forehead', 'redness_cheek', 'redness_forearm', 'redness_forehead']

```

```python
def minmax_normalize(df, features):
    for feature in features:
        df[feature] = (df[feature] - df[feature].min()) / (df[feature].max() - df[feature].min())
    return df
```

```python
import numpy as np
```

```python
# Remove outliers using the z-score method
df = minmax_normalize(df, features)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from two CSV files
df1 = pd.read_csv('/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/skinly_thailand_corrected/normalisation/min_max_thai.csv')
df2 = pd.read_csv('/home/azureuser/cloudfiles/code/Users/Akshayanivashini.ChandrasekarVijayalakshmi/skinly_thailand_corrected/outlier removal/cleaned_data_complete.csv')

# Identifying common numeric features in both dataframes
common_features = [col for col in df1.columns if col in df2.columns and df1[col].dtype in ['float64', 'int64']]

# Set up the plotting environment
sns.set(style="whitegrid")  # for a nicer grid style

# Iterate through each common feature and plot histograms
for feature in common_features:
    plt.figure(figsize=(12, 6))  # Set figure size for each feature comparison
    
    # Histogram for the first dataset
    plt.subplot(1, 2, 1)  # First subplot for the first dataset
    sns.histplot(df1[feature], color='blue', kde=True, element='step', label=f'Dataset 1 - {feature}')
    plt.title(f'Histogram of {feature} - Normalised in range 0 to 1')
    plt.xlabel(f'Values of {feature}')
    plt.ylabel('Frequency')
    plt.legend()
    
    # Histogram for the second dataset
    plt.subplot(1, 2, 2)  # Second subplot for the second dataset
    sns.histplot(df2[feature], color='green', kde=True, element='step', label=f'Dataset 2 - {feature}')
    plt.title(f'Histogram of {feature} - z score and custom')
    plt.xlabel(f'Values of {feature}')
    plt.ylabel('Frequency')
    plt.legend()
    
    plt.tight_layout()  # Adjust layout to prevent overlap of subplots
    plt.show()  # Display the plot
```
